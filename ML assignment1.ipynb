# Q2: Detect Outliers and Remove Them (Score: 4)

## a) Mean and Standard deviation


```python
# Define threshold
n = 3  # Number of standard deviations
mean = df['price_per_sqft'].mean()
std_dev = df['price_per_sqft'].std()
print('Mean:',mean)
print('std deviation:',std_dev)
```

```python
# Define boundaries
lower_limit = mean - n * std_dev
upper_limit = mean + n * std_dev
print("lower_limit:",lower_limit)
print("Upper limit:",upper_limit)
```

```python

# Remove outliers using trimming
df_trimmed = df[(df['price_per_sqft'] >= lower_limit) & (df['price_per_sqft'] <= upper_limit)]
df_trimmed
```

```python
# Remove outliers using capping
R=df['price_per_sqft'] = np.clip(df['price_per_sqft'], lower_limit, upper_limit)
R
```

# 4. Draw histplot to check the normality of the column(price per sqft column) and perform transformations if needed. Check the skewness and kurtosis before and after the transformation. (Score:1)

```python
import seaborn as sns
from scipy.stats import kurtosis, skew

# Original Data
original_data = df['price_per_sqft']

# Calculate Skewness and Kurtosis
original_skewness = skew(original_data)
original_kurtosis = kurtosis(original_data)

# Plot Histogram for Original Data
plt.figure(figsize=(12, 6))
sns.histplot(original_data, kde=True, bins=30, color='maroon')
plt.title(f'Original Data Distribution (Skewness: {original_skewness:.2f}, Kurtosis: {original_kurtosis:.2f})')
plt.xlabel('Price per Square Foot')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# Apply Log Transformation
transformed_data = np.log1p(original_data)  # Log transformation to handle zero/negative values

# Calculate Skewness and Kurtosis for Transformed Data
transformed_skewness = skew(transformed_data)
transformed_kurtosis = kurtosis(transformed_data)

# Plot Histogram for Transformed Data
plt.figure(figsize=(12, 6))
sns.histplot(transformed_data, kde=True, bins=30, color='orange')
plt.title(f'Transformed Data Distribution (Skewness: {transformed_skewness:.2f}, Kurtosis: {transformed_kurtosis:.2f})')
plt.xlabel('Log-Transformed Price per Square Foot')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# Print Skewness and Kurtosis Values
print(f"Original Data: Skewness = {original_skewness:.2f}, Kurtosis = {original_kurtosis:.2f}")
print(f"Transformed Data: Skewness = {transformed_skewness:.2f}, Kurtosis = {transformed_kurtosis:.2f}")
```
